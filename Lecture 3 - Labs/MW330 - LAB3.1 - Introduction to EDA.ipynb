```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("anscombe")
```

```python
df.head()
```

```python
sns.lmplot(x="x", y="y", col="dataset", data=df[df.dataset=='I'], ci=None)
plt.show()
```

```python
datasets = ['I', 'II', 'III', 'IV']
mean_x = {}
mean_y = {}
std_x = {}
std_y = {}

for ds in datasets:
    mean_x[ds] = round(df[df.dataset==ds].x.mean(), 2)
    std_x[ds] = round(df[df.dataset==ds].x.std(), 2)
    mean_y[ds] = round(df[df.dataset==ds].y.mean(), 2)
    std_y[ds] = round(df[df.dataset==ds].y.std(), 2)
    
print("mean (x): {}".format(mean_x))
print("std (x): {}".format(std_x))
print("mean (y): {}".format(mean_y))
print("std (x): {}".format(std_y))
```

```python
from sklearn import linear_model

datasets = ['I', 'II', 'III', 'IV']

for ds in datasets:
    regr = linear_model.LinearRegression()
    current_df = df[df.dataset==ds]  # slice to the current dataset I/II/III/IV
    regr.fit(current_df[['x']].values, current_df['y'])  # linear regression
    print("a = {:.2f}, b = {:.2f}".format(regr.intercept_, regr.coef_[0]))
    
```
