## Load and inspect dataset

```python
# dataset https://archive.ics.uci.edu/ml/datasets/auto+mpg

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

df = sns.load_dataset("mpg")

# in case you are working with an old seaborn version
# that has issues with NaNs:
# remove all nan-rows (which are only in the horsepower column)
#df[df.horsepower.isna()]
```

## Scatterplot

```python
sns.scatterplot(x=df.acceleration, y=df.mpg)
plt.show()
```

```python
#sns.scatterplot(x=df.acceleration, y=df.mpg, hue=df.origin)
sns.scatterplot(data=df, x='acceleration', y='mpg', hue='origin')
plt.show()
```

```python
#sns.scatterplot(x=df.acceleration, y=df.mpg, hue=df.origin, size=df.weight)
sns.scatterplot(data=df, x='acceleration', y='mpg', hue='origin', size='weight')
plt.show()
```

```python
sns.scatterplot(x=df.acceleration, y=df.mpg, hue=df.origin, size=df.weight, style=df.cylinders)
plt.show()
```

### Strip-Plot

```python
sns.scatterplot(x=df.origin, y=df.mpg)
plt.show()
```

```python
sns.stripplot(x="origin", y="mpg", data=df, jitter=0.2)
plt.show()
```

### Jointplot

```python
sns.jointplot('weight', 'mpg', data=df)
plt.show()
```

```python
sns.jointplot('weight', 'mpg', data=df, kind='kde')
plt.show()
```

```python
sns.jointplot('weight', 'mpg', data=df, kind='reg')
plt.show()
```

### Pairplot

```python
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df, hue='origin')
plt.show()
```

```python
sns.pairplot(df, kind='reg')
plt.show()
```
