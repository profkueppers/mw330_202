```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")
sns.pairplot(df, hue="species")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

all_features = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].values
all_classes = df['species'].values

dtc = DecisionTreeClassifier()
dtc.fit(all_features, all_classes)

# perform class predictions
print(dtc.predict([[5, 3.5, 1.5, 0.5]]))
print(dtc.predict([[7, 3.0, 5, 1.5]]))
print(dtc.predict([[7, 3.0, 4, 1.5]]))
```

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Separating out the features
X = df.loc[:, features].values

# Separating out the target
y = df.loc[:,['species']].values

# Standardizing the features
X = StandardScaler().fit_transform(X)

# Plotting the normalized data
iris_transformed = pd.DataFrame(X)
iris_transformed['species'] = y.copy()

sns.pairplot(iris_transformed, hue="species")
plt.show()
```

```python
df_copy = df.copy()
df_copy['sepal_area'] = df_copy.sepal_length * df_copy.sepal_width
df_copy['petal_area'] = df_copy.petal_length * df_copy.petal_width
del(df_copy['sepal_length'])
del(df_copy['sepal_width'])
del(df_copy['petal_length'])
del(df_copy['petal_width'])
sns.pairplot(df_copy, hue="species")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

principalComponents = pca.fit_transform(X)

principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])

principalDf.head()


# Plotting the automatically reduced data
iris_reduced = pd.DataFrame(principalDf)
iris_reduced['species'] = y.copy()

sns.pairplot(iris_reduced, hue="species")
plt.show()
```
