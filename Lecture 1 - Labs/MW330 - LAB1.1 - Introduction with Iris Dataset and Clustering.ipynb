```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
# let us put the samples into a dataframe called X
X = pd.DataFrame(load_iris().data)
X.tail()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(data=X)
plt.show()
```

```python
from sklearn.cluster import KMeans

no_clusters = 3

kmeans = KMeans(n_clusters = no_clusters)
y_kmeans = kmeans.fit_predict(X)

# visualized features
# 0=sepal_length, 1=sepal_width, 2=petal_length, 3=petal_width
feature1 = 0
feature2 = 1

# visualize the clusters
for cluster in range(0, no_clusters):
    plt.scatter(X[y_kmeans == cluster][feature1], X[y_kmeans == cluster][feature2])

# plot the centroids of each clusters
plt.scatter(kmeans.cluster_centers_[:, feature1], kmeans.cluster_centers_[:,feature2], 
            c = 'yellow')
plt.show()
```
