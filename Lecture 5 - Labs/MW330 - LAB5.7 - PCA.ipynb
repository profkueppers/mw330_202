```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = sns.load_dataset('diamonds')
df = df.iloc[:5000,:]
df['price_bin'] = pd.cut(df.price, 2)
```

# EDA

```python
sns.pairplot(data=df[['x','y','z']]);plt.show()
```

```python
sns.pairplot(data=df[['x', 'y', 'z', 'price_bin']], hue='price_bin');plt.show()
```

# PCA

```python
from sklearn.decomposition import PCA
X = df[['x', 'y', 'z']]  # these are the features we want to use
pca = PCA(n_components=2)
pca.fit(X)
```

```python
X_new = pd.DataFrame(pca.transform(X))
X_new['price_bin'] = df['price_bin']  # attach the original price bins (X_new has the same length as df!)
sns.pairplot(data=X_new, hue='price_bin');plt.show()
```
