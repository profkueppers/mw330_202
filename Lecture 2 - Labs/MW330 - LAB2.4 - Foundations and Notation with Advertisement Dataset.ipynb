```python
import seaborn as sns
import matplotlib.pyplot as plt

df['rounded_sales'] = round(df['sales'])
sns.pairplot(data=df.loc[:, ['TV', 'radio', 'newspaper', 'rounded_sales']], hue='rounded_sales')
plt.show()
```

```python
from sklearn import datasets, linear_model

X = df[['TV', 'radio', 'newspaper']]
y = df['sales']

# Create linear regression object
regr = linear_model.LinearRegression()

regr.fit(X, y)

# perform prediction (e.g. TV only) 
print(regr.predict([[200, 0, 0]]))
```
